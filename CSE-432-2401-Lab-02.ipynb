## Task 1
Generating simple Discrete Time Signal 

```python
import numpy as np
import matplotlib.pyplot as mpl
x = [2,1,0,1,0,3,0,2,0,0,0,1]
n = np.arange(-3,9,1)
mpl.stem(n,x)
```

## Task 2
Generating Continuous Time Signal 

```python
import numpy as np
import matplotlib.pyplot as mpl
n=np.arange(0,6*np.pi,0.01)
x=np.sin(n)
mpl.plot(n,x)
```

## Task 3
Decorating Plots

```python
import numpy as np
import matplotlib.pyplot as mpl
n=np.arange(0,6*np.pi,0.01)
x=np.sin(n)
y=np.cos(n)
mpl.plot(n,x, color='red')
mpl.plot(n,y, color='orange')
mpl.title("Sine and cosine Curve")
mpl.xlabel("n")
mpl.ylabel("Values")
mpl.show
```

## Task 4
Plotting multiple signals using subplots

```python
import numpy as np
import matplotlib.pyplot as mpl
n=np.arange(0,6*np.pi,0.01)
x=np.sin(n)
y=np.cos(n)
fig, ax = mpl.subplots(1, 2)
mpl.subplot(1,2,1)
mpl.plot(n,x, color='red')
mpl.title("sine Curve")
mpl.subplot(1,2,2)
mpl.plot(n,y, color='orange')
mpl.title("cosine Curve")
fig.tight_layout()
mpl.suptitle("Sinusoidal Graphs")
mpl.show
```

## Task 5
Observing the effect of sampling

```python
import numpy as np
import matplotlib.pyplot as mpl
fig, ax = mpl.subplots(3,1)
n=np.arange(0,6*np.pi,0.05)
x=np.sin(n)
mpl.subplot(3,1,1)
mpl.plot(n,x)
n=np.arange(0,6*np.pi,0.5)
y=np.sin(n)
mpl.subplot(3,1,2)
mpl.plot(n,y)
n=np.arange(0,6*np.pi,1)
x=np.sin(n)
mpl.subplot(3,1,3)
mpl.plot(n,x)
mpl.suptitle("Effect of Sampling")
fig.tight_layout()
mpl.show
```
