## Task 1
Signal time shifting

```python
import numpy as np
import matplotlib.pyplot as mpl
#Generating a DT signal
x = [2,1,0,1,0,3,0,2,0,0,0,1]
n = np.arange(-3,9,1)
mpl.stem(n,x)
mpl.title('Original Signal')

#Delay a signal by m points
m=2
n1 = np.arange(-3+m,9+m,1)
mpl.figure()
mpl.stem(n1,x)
mpl.title('Delay a signal by m points')

#advance a signal by m points
m=2
n2 = np.arange(-3-m,9-m,1)
mpl.figure()
mpl.stem(n2,x)
mpl.title('Advance a signal by m points')


mpl.figure()
fig, ax = mpl.subplots(3,1)
mpl.subplot(3,1,1)
mpl.stem(n,x)
mpl.title('Original Signal')
mpl.subplot(3,1,2)
mpl.stem(n1,x)
mpl.title('Delay a signal by m points')
mpl.subplot(3,1,3)
mpl.stem(n2,x)
mpl.title('Advance a signal by m points')
fig.tight_layout()
```

## Task 2
Folding operation on DT signal

```python
import numpy as np
import matplotlib.pyplot as mpl
#Generating a DT signal
x = [2,1,0,1,0,3,0,2,0,0,0,1]
n = np.arange(-3,9,1)
mpl.stem(n,x)
mpl.title('Original Signal')

#Folding
n1 = np.arange(-8,4,1)
x1=np.flip(x)
mpl.figure()
mpl.stem(n1,x1)
mpl.title('Folded Signal')
```
