## Task 1
Generating Sinusoidal Signal

```python
import numpy as np
import matplotlib.pyplot as mpl
F = 50  #Analog Frequency
Fs = 1000  #Sampling Rate
A=10  #Amplitude
theta=0 #initial phase
sampleSize=100
x = np.zeros(sampleSize)
size=np.arange(sampleSize)
for n in range(sampleSize):
    x[n] = A*np.sin(2*np.pi*F/Fs*n+theta)
mpl.stem(x)
#mpl.plot(size,x)
mpl.xlabel('Sample (n)')
mpl.ylabel('x [n]')
mpl.show()
```

## Task 2
Generating Unit Impulse Signal

```python
#Approach 1 symmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,6,1)
delta = np.zeros(len(n))
delta[len(n)//2]=1  #len(n)//2 is used to find  the mid poi9nt in the array. The operator // returns the floor of the division result
mpl.stem(n,delta)
mpl.xlabel('Sample (n)')
mpl.ylabel('$\delta [n]$')
mpl.title('Unit Impulse  ($\delta$) Signal')
mpl.show()
```

```python
#Approach 2 assymmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-50,11,1)
delta = np.zeros(len(n))
zeroPoint=0
for i in range(len(n)):
    if n[i]==0:
        zeroPoint=i
#print(zeroPoint)
delta[zeroPoint]=1  
mpl.stem(n,delta)
mpl.xlabel('Sample (n)')
mpl.ylabel('$\delta [n]$')
mpl.title('Unit Impulse  ($\delta$) Signal')
mpl.show()
```

## Task 3
Generating Unit Step Signal

```python
#Approach 1 symmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,6,1)
u = np.zeros(len(n))
for i in range(len(n)//2+1):
    u[len(n)//2+i]=1

mpl.stem(n,u)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u [n]$')
mpl.title('Unit Step  ($u$) Signal')
mpl.show()
```

```python
#Approach 2 assymmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,21,1)
u = np.zeros(len(n))
for i in range(len(n)):
    if n[i]>=0:
        u[i]=1
mpl.stem(n,u)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u [n]$')
mpl.title('Unit Step  ($u$) Signal')
mpl.show()
```

## Task 4
Generating Unit Ramp Signal

```python
#Approach 1 symmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,6,1)
u_r = np.zeros(len(n))
for i in range(len(n)//2+1):
    u_r[len(n)//2+i]=n[len(n)//2+i]

mpl.stem(n,u_r)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u_r [n]$')
mpl.title('Unit Ramp  ($u_r$) Signal')
mpl.show()
```

```python
#Approach 2 assymmetric
import numpy as np
import matplotlib.pyplot as mpl

n = np.arange(-5,21,1)
u_r = np.zeros(len(n))
for i in range(len(n)):
    if n[i]>=0:
        u_r[i]=n[i]
mpl.stem(n,u_r)
mpl.xlabel('Sample (n)')
mpl.ylabel('$u_r [n]$')
mpl.title('Unit Ramp  ($u_r$) Signal')
mpl.show()
```

## Task 5
Generating Exponential Signal

```python
import numpy as np
import matplotlib.pyplot as mpl
n = np.arange(0,20,1)
x = np.zeros(len(n))
a=1.25
for i in range (len(n)):
    x[i]=a**i
mpl.stem(n,x)
mpl.xlabel('Sample (n)')
mpl.ylabel('$x [n]$')
mpl.title('Exponential Signal (a=1.25)')
mpl.show()
```
