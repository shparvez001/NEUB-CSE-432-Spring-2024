## Task 1
Frequency Analysis of Discrete-Time Periodic Signals (Fourier Series)

```python
import numpy as np
import matplotlib.pyplot as mpl
#Generating Signal
#For generating periodic signal, the frequencies f2, f3 are harmonics (integer multiple) of f1.
f1=50
f2=100
f3=150
Fs=1000
L=256
N=512
x=np.zeros(N)
for i in range(L):
    x[i]=np.sin(np.pi*f1*i/Fs)+np.sin(np.pi*f2*i/Fs)+np.sin(np.pi*f3*i/Fs)
mpl.plot(x[0:L])

#Analysis
Ck=np.zeros(N)
for k in range(N):
    for n in range(N):
        Ck[k]=Ck[k]+x[n]*np.exp(-1j*2*np.pi*k*n/N)
mpl.figure()   
Y=np.abs(Ck[0:N//2])
mpl.plot(Y)

#Synthesis
x_r=np.zeros(N)
for n in range(N):
    for k in range (N):
        x_r[n]=x_r[n]+Ck[k]*np.exp(1j*2*np.pi*k*n/N)

mpl.figure()
mpl.plot(x_r[0:L])        
```
